```python
ccesdata = pd.read_csv('cces_example_data.csv')
ccesdata.head()
```

```python
ccesdata.describe()
```

```python
fig, (fstaxes, secaxes) = plt.subplots(1, 2)
fstaxes.set_title('Age Histogram')
secaxes.set_title('Political Identity')
plt.show()
```

```python
fig, axes = plt.subplots()
eval_count = ccesdata['Evaluation'].value_counts()
eval_count.plot(kind='bar', ax=axes)
axes.set_title('Evaluation - Frequency Bar Plot')
axes.set_xlabel('Evaluation')
axes.set_ylabel('Frequency')
plt.show()
```

```python
fig, axes = plt.subplots()
axes.scatter(ccesdata['ConMediaUse'], ccesdata['ImmiPhobia'], marker = 's')
axes.set_title('ConMediaUse vs ImmiPhobia')
axes.set_xlabel('ConMediaUse')
axes.set_ylabel('ImmiPhobia')
plt.show()
```

## Methods for improving *ConMediaUse vs ImmiPhobia* ScatterPlot
### For improving the ConMediaUse Vs ImmiPhobia scatterplot, Wecan consider the increasing marker transparency (alpha) to reveal overlapping points moreclearly (18.1). Adding jitter(18.1) to the data points can help distinguish similar values, and mapping(2.1) a third variable to colour will add a new dimension to the visualization. Adjusting (2.2) marker size or shape can further reduce overlap. Descriptive titles and axis labels, as suggested by matplotlib guidelines, it enhaces clarity.
### Finally, using colour gradients(2.3) based on Immiphobia intensity can make the patterns more visible.

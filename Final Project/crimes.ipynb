```python
import pandas as pd
```

```python
# Load the uploaded crime_data_cleanedset
file = 'Crime_Data_from_2020_to_Present.csv'
crime_data = pd.read_csv(file,low_memory=False)
crime_data.head()
```

```python
# 3. Define the list of columns to remove
columns_to_remove = [
    'AREA',
    'Part 1-2',
    'Crm Cd',
    'Mocodes',
    'Vict Descent',
    'Premis Cd',
    'Weapon Used Cd',
    'Status',
    'Crm Cd 1',
    'Crm Cd 2',
    'Crm Cd 3',
    'Crm Cd 4',
    'LOCATION',
    'Cross Street',
    'LAT',
    'LON',
    'TIME OCC'
]

# 4. Remove the specified columns
# Option 1: Create a new DataFrame without the unwanted columns
crime_data_cleaned = crime_data.drop(columns=columns_to_remove, errors='ignore')
```

```python
crime_data_cleaned.head()
```

```python
# Display a summary of the data to understand its structure
crime_data_cleaned.info()
crime_data_cleaned.shape
```

```python
crime_data_cleaned.isnull().sum()
```

```python
# Fill missing values
crime_data_cleaned['Vict Sex'].fillna('Unidentified') #, inplace=True
crime_data_cleaned['Weapon Desc'].fillna('None') #, inplace=True
```

```python
crime_data_cleaned.isnull().sum()
```

```python
# Converting 'DATE OCC' to datetime 
crime_data_cleaned['DATE OCC'] = pd.to_datetime(crime_data_cleaned['DATE OCC'], errors='coerce')
# onverting 'Date Rptd' is a datetime 
crime_data_cleaned['Date Rptd'] = pd.to_datetime(crime_data_cleaned['Date Rptd'], errors='coerce')


# Now split the 'Date Rptd' column into year, month, and day
crime_data_cleaned['Year Rptd'] = crime_data_cleaned['Date Rptd'].dt.year
crime_data_cleaned['Month Rptd'] = crime_data_cleaned['Date Rptd'].dt.month
crime_data_cleaned['Day Rptd'] = crime_data_cleaned['Date Rptd'].dt.day

```

```python

# Fill 'Date Rptd' with 'DATE OCC' if available
crime_data_cleaned['Date Rptd'].fillna(crime_data_cleaned['DATE OCC']) #, inplace=True

# Fill any remaining missing 'Date Rptd' with the earliest available date
earliest_date = crime_data_cleaned['Date Rptd'].min()
crime_data_cleaned['Date Rptd'].fillna(earliest_date) #, inplace=True

# Handling missing 'DATE OCC' values
#  Fill missing 'DATE OCC' with 'Date Rptd' where possible, or use a placeholder like the earliest available date in 'DATE OCC'

# Step 1: Fill 'DATE OCC' with 'Date Rptd' if available
crime_data_cleaned['DATE OCC'].fillna(crime_data_cleaned['Date Rptd'])#, inplace=True

# Step 2: Fill any remaining missing 'DATE OCC' with the earliest available date
earliest_date_occ = crime_data_cleaned['DATE OCC'].min()
crime_data_cleaned['DATE OCC'].fillna(earliest_date_occ)#, inplace=True
```
